```python
import os
import numpy as np
import nibabel as nib
import torch
import monai.transforms as transforms

import matplotlib.pyplot as plt

dataset_path = "/dhc/home/youngbin.ko/brain_data/train_for_guassian_120"
subfolders = os.listdir(dataset_path)

transform = transforms.Compose([transforms.ToTensor(dtype=torch.float)])

for subfolder in subfolders:
    print("Processing subfolder:", subfolder)

    flair_path = os.path.join(dataset_path, subfolder, subfolder + "_flair.nii.gz")
    t1_path = os.path.join(dataset_path, subfolder, subfolder + "_t1.nii.gz")
    t1ce_path = os.path.join(dataset_path, subfolder, subfolder + "_t1ce.nii.gz")
    t2_path = os.path.join(dataset_path, subfolder, subfolder + "_t2.nii.gz")
    print(flair_path, t1_path, t1ce_path, t2_path)

    flair_img = nib.load(flair_path)
    t1_img = nib.load(t1_path)
    t1ce_img = nib.load(t1ce_path)
    t2_img = nib.load(t2_path)

    flair_data = flair_img.get_fdata()
    t1_data = t1_img.get_fdata()
    t1ce_data = t1ce_img.get_fdata()
    t2_data = t2_img.get_fdata()
    
    imgs = [flair_data, t1_data, t1ce_data, t2_data]

    stacked_data = np.stack(imgs)

    transformed_data = transform(stacked_data)

    output_path = os.path.join(dataset_path, subfolder, subfolder + ".npy")

    np.save(output_path, transformed_data)

    print("Saved file:", output_path)
```
